## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# For other columns with missing values, we can impute or handle them based on the data characteristics and business context

# For example, we can impute missing values in 'reviews_per_month' with 0 since missing values may indicate no reviews

df['reviews_per_month'].fillna(0, inplace=True)
df['name'].fillna('Unknown', inplace=True)
df['host_name'].fillna('Unknown', inplace=True)

# Convert data types
df['last_review'] = pd.to_datetime(df['last_review'])

# Create new features if required
df['review_month'] = df['last_review'].dt.month
df['review_year'] = df['last_review'].dt.year
````

### What all manipulations have you done and insights you found?

1) We dropped rows with missing values in the 'price' column, as it is the target variable.

2) Imputed missing values in the 'reviews_per_month' column with 0.

3) changes the 'last_review' columns data type to Datetime.

4) Imputed missing values in the 'name' and 'host_name' columns with "Unknown".

**Insights:**

1) By dropping rows with missing values in the 'price' column, we ensured the integrity of our target variable for analysis.

2) Imputing missing values in the 'reviews_per_month' column with 0 allowed us to assume no reviews for listings with missing review data.

3) Imputing missing values in the 'name' and 'host_name' columns with "Unknown" maintained the integrity of these columns while handling missing data gracefully.

**Data Completeness:** The absence of null values and duplicates suggests that the dataset is well-maintained and of high quality, which increases confidence in the reliability of the analysis results.

**Further Analysis Opportunities:** The availability of clean and enriched data opens up opportunities for deeper analysis, such as investigating correlations between booking patterns and external factors like holidays or events in the city.

Overall, these manipulations have prepared the dataset for further exploratory analysis, enabling the extraction of valuable insights and trends related to Airbnb bookings in the dataset.

#### Chart - 8 : Pie Chart for Top Hosts by Listing Count

````python
# Chart - 8 visualization code
# Assuming 'calculated_host_listings_count' column represents the number of listings per host

# Calculate the number of listings per host
listings_per_host = df.groupby('host_name')['calculated_host_listings_count'].sum()

# Select only the top 10 hosts with the highest number of listings
top_10_hosts = listings_per_host.nlargest(10)

# Plotting the distribution of listings per host for the top 10 hosts using a pie chart
plt.figure(figsize=(8, 8))  # Adjust figure size to prevent overlapping
explode = [0.1] * len(top_10_hosts)  # Explode slices to make them stand out
plt.pie(top_10_hosts, labels=top_10_hosts.index, autopct='%1.1f%%', startangle=140, explode=explode)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Listings per Host (Top 10 Hosts)')
plt.show()
````

##### 1. Why did you pick the specific chart?

**Pie charts are suitable for displaying the composition of categorical data as parts of a whole. They provide a clear visual representation of proportions.**
